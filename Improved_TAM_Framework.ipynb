```python
import matplotlib.pyplot as plt
import networkx as nx

# Create a directed graph
G = nx.DiGraph()

# Define nodes (TAM constructs)
nodes = ["AI Literacy", "Trust", "Perceived Usefulness (PU)",
         "Perceived Ease of Use (PEOU)", "Emotion", "Behavioral Intention (BI)"]

G.add_nodes_from(nodes)

# Define edges with weights (regression coefficients from cited studies)
edges = [
    ("AI Literacy", "Perceived Ease of Use (PEOU)", 0.367),      # Vo & Nguyen (2024)
    ("Trust", "Perceived Usefulness (PU)", 0.412),              # Tan et al. (2024)
    ("Perceived Usefulness (PU)", "Behavioral Intention (BI)", 0.41),                  # Example strong path (Tan et al. 2024)
    ("Perceived Ease of Use (PEOU)", "Perceived Usefulness (PU)", 0.35),                # Classic TAM link
    ("Perceived Ease of Use (PEOU)", "Behavioral Intention (BI)", 0.28),                # Commonly significant
    ("Emotion", "Behavioral Intention (BI)", -0.25)             # Ali et al. (2023) - negative influence
]

# Add edges with weights
for source, target, weight in edges:
    G.add_edge(source, target, weight=weight)

# Node positions (manual for better readability)
pos = {
    "AI Literacy": (-2, 0),
    "Trust": (-2, -1.5),
    "Perceived Ease of Use (PEOU)": (0, 1),
    "Perceived Usefulness (PU)": (0, -1),
    "Emotion": (0, -3),
    "Behavioral Intention (BI)": (2, -1)
}

# Define brighter node colors
node_colors = ["#ed0909", "#0a1bcf", "#0bb50b", "#0bb50b", "#B266FF", "#f7ef07"]

# Draw nodes
plt.figure(figsize=(10, 7))
nx.draw_networkx_nodes(G, pos, node_size=2500, node_color=node_colors, edgecolors="black")

# Draw labels
nx.draw_networkx_labels(G, pos, font_size=9, font_weight="bold")

# Draw edges with arrows
edges_drawn = nx.draw_networkx_edges(G, pos, arrowstyle="->", arrowsize=20, edge_color="gray")

# Add edge labels (coefficients)
edge_labels = {(u, v): f"β={d['weight']:.2f}" for u, v, d in G.edges(data=True)}
nx.draw_networkx_edge_labels(G, pos, edge_labels=edge_labels, font_size=8, font_color="darkred")

plt.title("Expanded TAM for AI Adoption in SME Accounting (2023–2024 Studies)", fontsize=12, fontweight="bold")
plt.axis("off")
plt.show()
```

```python
print(f"Number of nodes: {len(G.nodes())}")
print(f"Number of edges: {len(G.edges())}")
print(f"Is the graph weakly connected: {nx.is_weakly_connected(G)}")
```

## Analyze node centrality

### Subtask:
Calculate centrality measures (e.g., degree centrality, betweenness centrality) to understand the importance of each node in the network.


**Reasoning**:
Calculate and store the centrality measures for each node.



```python
in_degree_centrality = nx.in_degree_centrality(G)
out_degree_centrality = nx.out_degree_centrality(G)
betweenness_centrality = nx.betweenness_centrality(G)

centrality_data = {
    'In-Degree Centrality': in_degree_centrality,
    'Out-Degree Centrality': out_degree_centrality,
    'Betweenness Centrality': betweenness_centrality
}

import pandas as pd
centrality_df = pd.DataFrame(centrality_data)
display(centrality_df)
```

## Analyze edge properties

### Subtask:
Examine edge weights and directions to understand the strength and direction of the relationships between constructs.


**Reasoning**:
Iterate through the edges and print the source, target, direction, and weight of each edge to understand the relationships between constructs.



```python
print("Edge Properties (Strength and Direction of Relationships):")
for u, v, data in G.edges(data=True):
    weight = data['weight']
    print(f"  From: {u} -> To: {v}, Weight (Strength): {weight:.3f}")
```

## Identify key paths

### Subtask:
Explore paths between important nodes to understand the flow of influence in the model.


**Reasoning**:
Identify source and target nodes, find all simple paths between them, and calculate the cumulative weight for each path.



```python
# Identify key source nodes (high out-degree, low in-degree)
source_nodes = ["AI Literacy", "Trust", "Emotion"]

# Identify the key target node (highest in-degree)
target_node = "Behavioral Intention (BI)"

print(f"Exploring paths from source nodes {source_nodes} to target node '{target_node}':\n")

# Find and print all simple paths and their cumulative weights
for source_node in source_nodes:
    print(f"Paths from '{source_node}' to '{target_node}':")
    try:
        # Find all simple paths between source and target
        for path in nx.all_simple_paths(G, source=source_node, target=target_node):
            cumulative_weight = 0
            path_edges = []
            # Calculate cumulative weight along the path
            for i in range(len(path) - 1):
                u, v = path[i], path[i+1]
                weight = G[u][v]['weight']
                cumulative_weight += weight
                path_edges.append(f"{u} -> {v} (β={weight:.2f})")
            print(f"  Path: {' -> '.join(path)}")
            print(f"  Edges: {', '.join(path_edges)}")
            print(f"  Cumulative Weight: {cumulative_weight:.3f}\n")
    except nx.NetworkXNoPath:
        print(f"  No path exists from '{source_node}' to '{target_node}'.\n")
```

Present the results of the network analysis in a clear and understandable way.
